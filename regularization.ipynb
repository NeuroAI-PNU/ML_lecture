```python
"""
L1 Regularization 시각화
========================
L = (w-5)^2,  L1 = λ|w|,  L_reg = L + λ|w|

슬라이더로 λ 값을 조절하며 최적점 이동을 관찰할 수 있습니다.
"""

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display


# ── 손실 함수 정의 ──────────────────────────────────────
def loss_L(w):
    return (w - 5) ** 2

def loss_L1(w, lam):
    return lam * np.abs(w)

def loss_L1reg(w, lam):
    return loss_L(w) + loss_L1(w, lam)

def find_w_star(lam):
    if lam >= 10:
        return 0.0
    else:
        return 5.0 - lam / 2.0

w = np.linspace(-3, 10, 500)

def plot_l1(lam=1.0):
    fig, ax = plt.subplots(figsize=(10, 6))

    ax.plot(w, loss_L(w),             color='#378ADD', lw=2.5, label=r'$L = (w-5)^2$')
    ax.plot(w, loss_L1(w, lam),       color='#1D9E75', lw=2,   ls='--', label=r'$\lambda\,|w|$')
    ax.plot(w, loss_L1reg(w, lam),      color='#D85A30', lw=2.5, label=r'$L_{\mathrm{reg}} = L + \lambda\,|w|$')

    ws = find_w_star(lam)
    ax.plot(ws, loss_L1reg(ws, lam), 'o', ms=10, mec='#D85A30', mfc='white', mew=2.5,
            label=r'$\min(L_{\mathrm{reg}})$', zorder=5)
    ax.plot(5, 0, 'o', ms=10, mec='#378ADD', mfc='white', mew=2.5,
            label=r'$\min(L),\; w^*\!=5$', zorder=5)

    ax.annotate('', xy=(ws, 4), xytext=(5, 4),
                arrowprops=dict(arrowstyle='<->', color='#888', lw=1.5))

    shift = 5.0 - ws
    ax.text(0.5, 0.95,
            f'λ = {lam:.1f}  →  w* = {ws:.1f}  (move {shift:.1f})',
            transform=ax.transAxes, ha='center', va='top', fontsize=12, color='#444',
            bbox=dict(boxstyle='round,pad=0.4', fc='#FFF8F0', ec='#D85A30', alpha=0.9))

    y_max = max(80, loss_L1reg(w[0], lam) * 1.1)
    ax.set_xlim(-3, 10)
    ax.set_ylim(0, y_max)
    ax.set_xlabel('w', fontsize=13)
    ax.set_ylabel('Loss', fontsize=13)
    ax.set_title('L1 Regularization', fontsize=15, fontweight='bold')
    ax.legend(loc='upper left', fontsize=10, framealpha=0.9)
    ax.axvline(x=0, color='gray', lw=0.5, alpha=0.3)
    ax.grid(True, alpha=0.15)
    plt.tight_layout()
    plt.show()

slider = widgets.FloatSlider(value=1.0, min=0, max=15, step=0.1, description='λ',
                             continuous_update=True, layout=widgets.Layout(width='70%'))
widgets.interact(plot_l1, lam=slider)
```

```python
"""
L2 Regularization 시각화
========================
L = (w-5)^2,  L2 = λw^2,  L_reg = L + λw^2

슬라이더로 λ 값을 조절하며 최적점 이동을 관찰할 수 있습니다.

L1과의 차이점:
  - L2는 w*가 절대 0이 되지 않고 0으로 수렴만 함 (soft shrinkage)
  - w* = 5 / (1 + λ)
"""


# ── 손실 함수 정의 ──────────────────────────────────────
def loss_L2(w, lam):
    return lam * w ** 2

def loss_L2reg(w, lam):
    return loss_L(w) + loss_L2(w, lam)

def find_w_star(lam):
    """
    L_reg = (w-5)^2 + λw^2 의 최솟값.
    dL/dw = 2(w-5) + 2λw = 0  →  w* = 5 / (1 + λ)
    λ→∞ 일수록 w*→0 이지만 절대 0이 되지는 않음 (L2의 특성).
    """
    return 5.0 / (1.0 + lam)

w = np.linspace(-3, 10, 500)

def plot_l2(lam=1.0):
    fig, ax = plt.subplots(figsize=(10, 6))

    ax.plot(w, loss_L(w),        color='#378ADD', lw=2.5, label=r'$L = (w-5)^2$')
    ax.plot(w, loss_L2(w, lam),  color='#1D9E75', lw=2,   ls='--', label=r'$\lambda\,w^2$')
    ax.plot(w, loss_L2reg(w, lam), color='#D85A30', lw=2.5, label=r'$L_{\mathrm{reg}} = L + \lambda\,w^2$')

    ws = find_w_star(lam)
    ax.plot(ws, loss_L2reg(ws, lam), 'o', ms=10, mec='#D85A30', mfc='white', mew=2.5,
            label=r'$\min(L_{\mathrm{reg}})$', zorder=5)
    ax.plot(5, 0, 'o', ms=10, mec='#378ADD', mfc='white', mew=2.5,
            label=r'$\min(L),\; w^*\!=5$', zorder=5)

    ax.annotate('', xy=(ws, 4), xytext=(5, 4),
                arrowprops=dict(arrowstyle='<->', color='#888', lw=1.5))

    shift = 5.0 - ws
    ax.text(0.5, 0.95,
            f'λ = {lam:.1f}  →  w* = {ws:.2f}  (move {shift:.2f})',
            transform=ax.transAxes, ha='center', va='top', fontsize=12, color='#444',
            bbox=dict(boxstyle='round,pad=0.4', fc='#F0F8FF', ec='#378ADD', alpha=0.9))

    y_max = max(80, loss_L2reg(w[0], lam) * 1.1)
    ax.set_xlim(-3, 10)
    ax.set_ylim(0, y_max)
    ax.set_xlabel('w', fontsize=13)
    ax.set_ylabel('Loss', fontsize=13)
    ax.set_title('L2 Regularization', fontsize=15, fontweight='bold')
    ax.legend(loc='upper left', fontsize=10, framealpha=0.9)
    ax.axvline(x=0, color='gray', lw=0.5, alpha=0.3)
    ax.grid(True, alpha=0.15)
    plt.tight_layout()
    plt.show()

slider = widgets.FloatSlider(value=1.0, min=0, max=15, step=0.1, description='λ',
                             continuous_update=True, layout=widgets.Layout(width='70%'))
widgets.interact(plot_l2, lam=slider)
```
